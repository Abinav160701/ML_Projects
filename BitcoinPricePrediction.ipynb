```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
```

```python
df=pd.read_csv("bitcoin.csv")
```

```python
df.head()
```

```python
df.drop(["Date"],1,inplace=True)
```

```python
df.head()
```

```python
predday=30
df["predictions"]=df['Price'].shift(-predday)
df.head()
```

```python
df.tail()
```

```python
X=np.array(df.drop(['predictions'],axis=1))
```

```python
X=X[:len(df)-predday]
```

```python
print(X)
```

```python
y=np.array(df['predictions'])
y=y[:-predday]
print(y)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.2)
predday_array=np.array(df.drop(["predictions"],1))[-predday:]
print(predday_array)
```

```python
xtrain.shape
```

```python
ytrain.shape
```

```python
## SVR
svr_rbf = SVR(kernel='poly', C=1e3, gamma=0.00001)
svr_rbf.fit(xtrain, ytrain)
```

```python
## testing
svr_rbf_confidence = svr_rbf.score(xtest,ytest)
print('SVR_RBF accuracy :',svr_rbf_confidence)
```

```python
# print the predicted values
svm_prediction = svr_rbf.predict(xtest)
print(svm_prediction)
print()
print(ytest)
```

```python
from sklearn.metrics import mean_squared_error
import math
MSE = mean_squared_error(ytest, svm_prediction)
 
RMSE = math.sqrt(MSE)
print("Root Mean Square Error:\n")
print(RMSE)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.scatter(ytest, svm_prediction, c='crimson')
plt.yscale('log')
plt.xscale('log')

p1 = max(max(svm_prediction), max(ytest))
p2 = min(min(svm_prediction), min(ytest))
plt.plot([p1, p2], [p1, p2], 'b-')
plt.xlabel('True Values', fontsize=15)
plt.ylabel('Predictions', fontsize=15)
plt.axis('equal')
plt.show()
```

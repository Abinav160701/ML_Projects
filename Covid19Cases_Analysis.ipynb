```python
import pandas as pd
```

```python
df = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
```

```python
df.head()
```

```python
df = df.iloc[:-1, 1:]
```

```python
df[df.state_name=='Delhi']
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
total_states = np.arange(len(df['state_name']))
```

```python
total_states
```

```python
df['active'].sort_values(ascending=False)
```

## Total Active Cases Based On States in India

```python
plt.figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')
# plt.figure(figsize=(9, 6))

plt.barh(total_states, df['active'], align='center', alpha=0.5, 
        color=(1, 0, 0), edgecolor=(0.5, 0.2, 0.8))
plt.yticks(total_states,df['state_name'])
plt.xlim(1, max(df['active'])+10)
plt.xlabel('Active Number of Cases')
plt.title("Corona Virus Cases")
plt.show()
```

## Positive Number of Cases based on States

```python
plt.figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')
# plt.figure(figsize=(9, 6))

plt.barh(total_states, df['positive'], align='center', alpha=0.5, 
        color=(1, 0, 0), edgecolor=(0.5, 0.2, 0.8))
plt.yticks(total_states,df['state_name'])
plt.xlim(1, max(df['positive'])+100000)
plt.xlabel('Postive Number of Cases')
plt.title("Corona Virus Cases")
plt.show()
```

```python
plt.figure(num=None, figsize=(9, 6), dpi=80, facecolor='w', edgecolor='k')
# plt.figure(figsize=(9, 6))

plt.barh(total_states, df['death'], align='center', alpha=0.5, 
        color=(0, 0, 1), edgecolor=(0.5, 0.2, 0.8))
plt.yticks(total_states,df['state_name'])
plt.xlim(1, max(df['death'])+100)
plt.xlabel('Death Number of Cases')
plt.title("Corona Virus Cases")
plt.show()
```
